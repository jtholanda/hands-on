### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
# Cria uma nova coluna "MES" a partir da coluna "DATA FINAL", extraindo apenas o número do mês.
df["MES"] = df["DATA FINAL"].apply(lambda x: x.month)

# Filtra o DataFrame original para considerar apenas as linhas onde o estado é "RIO DE JANEIRO".
df_rio = df[df["ESTADO"] == "RIO DE JANEIRO"]

# Exibe o DataFrame filtrado contendo apenas os dados do Rio de Janeiro.
df_rio.sort_values(by='ANO-MES').head(15)
```

```python
# Agrupa os dados por "ANO-MES" e calcula a média do "PREÇO MÉDIO REVENDA" para cada grupo.
df_rio.groupby("ANO-MES")[["PREÇO MÉDIO REVENDA"]].mean()
```

```python
# Agrupa por "ANO-MES" e pega o último valor registrado no mês para "PREÇO MÉDIO REVENDA".
df_rio.groupby(["ANO-MES"])[["PREÇO MÉDIO REVENDA"]].last()
```

```python
# Agrupa por "ANO-MES", pegando o último valor do mês para as colunas "PREÇO MÉDIO REVENDA" e "MES".
# Isso garante que, para cada "ANO-MES", temos o preço final do mês e o número do mês.
df_month_rio = df_rio.groupby("ANO-MES")[["PREÇO MÉDIO REVENDA", "MES"]].last()
df_month_rio
```

```python
df_month_rio[df_month_rio["MES"] == 12]
```

```python
df_month_rio[df_month_rio["MES"] == 12].shift(1)
```

```python
df_month_rio[df_month_rio["MES"] == 12] / df_month_rio[df_month_rio["MES"] == 12].shift(1)
```

```python
# Calcula a variação percentual ano a ano apenas para o mês de dezembro.
# 1. Filtra linhas onde "MES" == 12 (dezembro)
# 2. Divide os valores de dezembro de um ano pelo valor de dezembro do ano anterior (.shift(1))
# 3. Subtrai 1 para obter a variação relativa
# 4. Multiplica por 100 para converter em percentual
(df_month_rio[df_month_rio["MES"] == 12] / df_month_rio[df_month_rio["MES"] == 12].shift(1) - 1) * 100
```

```python
df_serie_calculate['DIFERENCA_ABSOLUTA'] = abs(df_serie_calculate["PRIMEIRO"]-df_serie_calculate["ÚLTIMO"])
df_serie_calculate
```

```python
# Agrupa o DataFrame por "ANO-MES" e pega o valor máximo da coluna "PREÇO MÉDIO REVENDA" de cada mês.
# Depois, garante que a série resultante esteja ordenada cronologicamente pelo índice (ANO-MES).
df_max = df.groupby("ANO-MES").max()["PREÇO MÉDIO REVENDA"].sort_index()

# Dentro de cada grupo por "ANO-MES", encontra o índice (linha) onde o preço de revenda foi o MAIOR.
# Isso é útil para, por exemplo, recuperar todas as colunas dessa linha posteriormente.
idx_max = df.groupby('ANO-MES')["PREÇO MÉDIO REVENDA"].idxmax()

# Agrupa por "ANO-MES" e pega o valor mínimo da coluna "PREÇO MÉDIO REVENDA" de cada mês.
# Ordena o resultado para manter a sequência cronológica.
df_min = df.groupby("ANO-MES").min()["PREÇO MÉDIO REVENDA"].sort_index()

# Dentro de cada grupo por "ANO-MES", encontra o índice (linha) onde o preço de revenda foi o MENOR.
idx_min = df.groupby('ANO-MES')["PREÇO MÉDIO REVENDA"].idxmin()

# Cria um DataFrame vazio que provavelmente será usado para armazenar a diferença entre valores máximos e mínimos.
df_diff = pd.DataFrame()
```

```python
idx_max
```

```python
# Calcula a diferença absoluta entre o valor máximo e o valor mínimo
# Obs.: Aqui está usando uma simples subtração, sem aplicar valor absoluto (abs()), 
#       então o resultado pode ser negativo se df_max < df_min.
df_diff["abs_diff"] = df_max - df_min  

# Calcula a diferença percentual em relação ao valor mínimo
# Fórmula: ((valor_max - valor_min) / valor_min) * 100
# Isso mostra quanto o valor máximo está acima do mínimo, em termos percentuais.
df_diff["perct_diff"] = (df_max - df_min) / df_min * 100  

# Armazena o valor máximo encontrado
df_diff["max"] = df_max  

# Armazena o valor mínimo encontrado
df_diff["min"] = df_min  
```

```python
# Seleciona do DataFrame original `df` as linhas correspondentes a `idx_max`
# e exibe apenas as colunas "ESTADO", "PREÇO MÉDIO REVENDA" e "ANO-MES".
df.loc[idx_max, :][["ESTADO", "PREÇO MÉDIO REVENDA", "ANO-MES"]]  
```

```python
# Preenche a coluna "ESTADO_MAX" de df_diff com os nomes dos estados
# correspondentes aos índices onde ocorreu o valor máximo em df
df_diff["ESTADO_MAX"] = df.loc[idx_max, :]["ESTADO"].values

# Preenche a coluna "ESTADO_MIN" de df_diff com os nomes dos estados
# correspondentes aos índices onde ocorreu o valor mínimo em df
df_diff["ESTADO_MIN"] = df.loc[idx_min, :]["ESTADO"].values
```
